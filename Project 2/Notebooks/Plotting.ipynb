```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np



sns.set(style="darkgrid")
sns.set(font_scale=1.3)

np.random.seed(42)
```

```python
df = pd.read_csv('Final.csv')
```

```python
sns.catplot(x="class", data=df, kind="count", height=6, aspect=1.5)
plt.ylabel('Tweets Number')
plt.xlabel('Polarity Class')
plt.title('Sentiment Analysis with ENDSARS Tweets')
plt.show()
```

```python
f, ax = plt.subplots(figsize=(12,8))
colors = {'negative':'red', 'neutral':'blue', 'positive':'green'}      
labels = list(colors.keys())
handles = [plt.Rectangle((0,0),1,1, color=colors[label]) for label in labels]

plt.bar([1535,465,923], [1,0.5,0.8], width=[150,150,150], bottom=[-1.0,0.0,0.2], color=['red','blue','green'])
plt.ylabel('Sentiment Polarity')
plt.xlabel('Number of Tweets')
plt.title('Sentiment Analysis with ENDSARS Tweets')
plt.legend(handles, labels)
plt.show()
```

```python
y=[i for i in range(30)]
x=np.random.ranf(size=30)
f, ax = plt.subplots(figsize=(12,8))
plt.plot(x,y, label='Polarity Details!', marker='o')
plt.xlabel('Sentiment Subjectivity')
plt.ylabel('Tweets Number')
plt.title('Sentiment Analysis with ENDSARS Tweets')
plt.legend()
plt.show()
```
