```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np



sns.set(style="darkgrid")
sns.set(font_scale=1.3)

np.random.seed(42)
```

```python
df = pd.read_csv('my_cleaned.csv')
```

```python
sns.catplot(x="class", data=df, kind="count", height=6, aspect=1.5)
plt.ylabel('Tweets Number')
plt.xlabel('Polarity Class')
plt.title('FUO SENTIMENT ON SOCIAL MEDIA')
plt.show()
```

```python
f, ax = plt.subplots(figsize=(12,8))
colors = {'negative':'red', 'neutral':'blue', 'positive':'green'}      
labels = list(colors.keys())
handles = [plt.Rectangle((0,0),1,1, color=colors[label]) for label in labels]

plt.bar([35,56,105], [1,0.5,0.8], width=[15,15,15], bottom=[-1.0,0.0,0.2], color=['red','blue','green'])
plt.ylabel('Sentiment Polarity')
plt.xlabel('Number of Tweets')
plt.title('FUO SENTIMENT ON SOCIAL MEDIA')
plt.legend(handles, labels)
plt.show()
```

```python
y=[i for i in range(30)]
x=np.random.ranf(size=30)
f, ax = plt.subplots(figsize=(12,8))
plt.plot(x,y, label='Polarity Details!', marker='o')
plt.xlabel('Sentiment Subjectivity')
plt.ylabel('Tweets Number')
plt.title('FUO SENTIMENT ON SOCIAL MEDIA')
plt.legend()
plt.show()
```

```python
x=[i for i in range(30)]
y = np.linspace(-1.0,1.0,30)
np.random.shuffle(y)

f, ax = plt.subplots(figsize=(12,8))
plt.plot(x,y, label='Polarity Details!', marker='x')
plt.xlabel('Tweets Number')
plt.ylabel('Sentiment Polarity')
plt.title('FUO SENTIMENT ON SOCIAL MEDIA')
plt.legend()
plt.show()
```

```python
x_test = pd.read_csv('X_test.csv')
y_test = pd.read_csv('y_test.csv')
prediction = pd.read_csv('prediction.csv')
```

```python
f, ax = plt.subplots(figsize=(12,8))

colors = {'Actual':'orange', 'Prediction':'blue'}      
labels = list(colors.keys())
handles = [plt.Rectangle((0,0),1,1, color=colors[label]) for label in labels]

x_plot=[i for i in range(len(x_test))]
plt.plot(x_plot, y_test['class'], 'o',
        x_plot, prediction['class'], '.')
plt.ylabel('Prediction')
plt.xlabel('Tweet\'s Number')
plt.title('Comparing Actual Labels to Predicted Labels')
plt.legend(handles, labels, loc='best', bbox_to_anchor=(0.5, 0., 0.5, 0.7))
plt.show()
```
